```python
class VAE(nn.Module):
    def __init__(self, T, n_x, n_z):
        super().__init__()
        
        mult = 120
        
        self.enc = nn.Sequential(
            nn.Linear(T*n_x, mult*5),
            nn.BatchNorm1d(mult*5),
            nn.ELU(),
            
            nn.Linear(mult*5, mult*3),
            nn.BatchNorm1d(mult*3),
            nn.ELU(),
            
            nn.Linear(mult*3, mult*2),
            nn.BatchNorm1d(mult*2),
            nn.ELU(),
            
            Latent(mult*2, n_z)
        )
        
        
        self.dec = nn.Sequential(
            nn.Linear(n_z, mult*2),
            nn.BatchNorm1d(mult*2),
            nn.ELU(),
            
            nn.Linear(mult*2, mult*3),
            nn.BatchNorm1d(mult*3),
            nn.ELU(),
            
            
            nn.Linear(mult*3, mult*5),
            nn.BatchNorm1d(mult*5),
            nn.ELU(),
            
            Normal(mult*5, T*n_x)
        )
    
    def _dkl_loss(self, mu, log_var):
        kld = -0.5 * torch.sum(1 + log_var - mu.pow(2) - log_var.exp(), dim=1)
        return kld
    
    def _rcl_loss(self, dist, x):
        return -dist.log_prob(x).sum(dim=1)
    
    def encode(self, x):
        return self.enc(x)

    def decode(self, z):
        return self.dec(z)
    
    def forward(self, x, beta = 1.0):
        z, mu, log_var = self.encode(x)
        dkl = self._dkl_loss(mu, log_var)
        dist = self.decode(z)
        rcl = self._rcl_loss(dist, x)
        #print(rcl.mean(), dkl.mean())
        return dist, torch.mean(rcl + dkl*beta)
```

```python
class ConvNorm(nn.Module):
    def __init__(self, i, o):
        super().__init__()
        self.mu = nn.Conv1d(in_channels = i, out_channels = o, kernel_size = 5, stride=1, padding=2)
        self.log_var = nn.Conv1d(in_channels = i, out_channels = o, kernel_size = 5, stride=1, padding=2)
    
    def forward(self, x):
        mu = self.mu(x)
        log_var = self.log_var(x)
        sigma = torch.exp(0.5 * log_var) + 1e-4
        return td.LogNormal(mu, sigma)
```

```python
class ConvVAE(nn.Module):
    def __init__(self, T, n_x, n_z):
        super().__init__()
        self.T = T
        self.n_x = n_x
        self.n_z = n_z
        mult = 5
        
        self.enc = nn.Sequential(
            nn.Conv1d(in_channels = n_x, out_channels = mult, kernel_size = 5, stride=1, padding=2),
            nn.BatchNorm1d(mult),
            nn.ELU(),
            
            nn.Conv1d(in_channels = mult, out_channels = mult*2, kernel_size = 5, stride=2, padding=2),
            nn.BatchNorm1d(mult*2),
            nn.ELU(),
            
            nn.Flatten(),
            
            nn.Linear(310, 150),
            nn.BatchNorm1d(150),
            nn.ELU(),
            
            Latent(150, n_z)
        )
        self.dec = nn.Sequential(
            nn.Linear(n_z, 150),
            nn.BatchNorm1d(150),
            nn.ELU(),
            
            nn.Linear(150, 310),
            nn.BatchNorm1d(310),
            nn.ELU(),
            
            Reshape(10, 31),
            
            nn.ConvTranspose1d(in_channels = mult*2, out_channels = mult, kernel_size = 5, stride=2, padding=2),
            nn.BatchNorm1d(mult),
            nn.ELU(),
            
            ConvNorm(mult, n_x)
        )
    def _dkl_loss(self, mu, log_var):
        kld = -0.5 * torch.sum(1 + log_var - mu.pow(2) - log_var.exp(), dim=1)
        return kld
    
    def _rcl_loss(self, dist, x):
        x = x.view(-1, self.n_x, self.T)
        return -dist.log_prob(x).sum(dim=1).sum(dim=1)
    
    def encode(self, x):
        #print(x.shape)
        x = x.view(-1, self.n_x, self.T)
        return self.enc(x)

    def decode(self, z):
        return self.dec(z)
    
    def forward(self, x, beta = 1.0):
        z, mu, log_var = self.encode(x)
        dkl = self._dkl_loss(mu, log_var)
        dist = self.decode(z)
        rcl = self._rcl_loss(dist, x)
        return dist, torch.mean(rcl + dkl*beta)
```

```python
def gradmvn(x, dist):
    mu = dist.mean
    Sigma = dist.covariance_matrix + 0.02*torch.ones(x.size(0), x.size(1)).diag_embed()
    Sigma_inv = torch.inverse(Sigma)
    bb = (x - mu).unsqueeze(2)
    #print(bb.shape, Sigma_inv.shape)
    return bb.transpose(1,2).bmm(Sigma_inv).squeeze(1)

def gradt(x, dist):
    df = dist.df
    return x * (df + 1) / (df + x**2)

def grad(x, dist):
    
    mu = dist.mean
    sigma = dist.scale
    grad = (x - mu) / sigma.pow(2)
    return grad
    

def sgd(x, vae, N=200, alpha=1, eps = 1):
    z_list = []
    x_list = []
    #x = x.view(-1, n_x, T)
    for i in range(N):
        eps = eps / np.sqrt(i+1)
        x_z = x + np.sqrt(eps) * torch.randn_like(x)
        #print(x_z.shape, "x")
        z, mu, log_var = vae.encode(x_z)
        dist = vae.decode(z)
        z_list.append(mu.unsqueeze(1))
        g = grad(x_z, dist)
        #g /= g.norm()
        #print(g.shape, "G")
        x = x - alpha * g / np.sqrt(i+1)
        x_list.append(x.flatten(1).unsqueeze(1))
    return torch.cat(z_list, 1), torch.cat(x_list ,1)
```
